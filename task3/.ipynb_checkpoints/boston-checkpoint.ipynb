```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df_x = pd.DataFrame(boston.data , columns=boston.feature_names)
df_y = pd.DataFrame(boston.target,columns=['Cost * 10,000$'])
```

```python
df_x
```

```python
df_y
```

```python
df_x.info()
```

```python
xTrain,xTest,y_Train,y_Test = train_test_split(df_x,df_y,test_size=0.2)
```

```python
trainData = xTrain.join(y_Train)
testData = xTest.join(y_Test)
```

```python
trainData.hist(figsize=(20,20))
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(trainData.corr(),annot=True,cmap="RdBu",vmax=1,vmin=-1)
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="LSTAT",y="RM",data=trainData,hue="Cost * 10,000$",palette="coolwarm",s=100)
```

```python
reg = LinearRegression()
regressor = reg.fit(xTrain,y_Train)
```

```python
#not bad and not good
regressor.score(xTest,y_Test)
```

```python
#lets tray Random Forest
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
Forest = forest.fit(xTrain,y_Train)
```

```python
Forest.score(xTest,y_Test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model_dt = DecisionTreeRegressor(max_depth=5, random_state=42)
```

```python
model_dt.fit(xTrain, y_Train)
```

```python
 model_dt.score(xTest, y_Test)
```

```python
from xgboost import XGBRegressor
```

```python
model_xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
```

```python
model_xgb.fit(xTrain, y_Train)
```

```python
 model_xgb.score(xTest, y_Test)
```

```python
arr = [1.0, 12.0, 20.0, 0.0, 0.5, 6.0, 50.0, 8.0, 5.0, 300.0, 15.0, 400.0, 20.0]
new_data = pd.DataFrame([arr], columns=boston.feature_names)
y_pred = model_xgb.predict(new_data)
```
